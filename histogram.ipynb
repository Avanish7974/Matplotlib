```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(size=1000)
plt.hist(data,bins=10,color = 'blue')
plt.show()
```

```python
labele = ['apple', 'banana', 'orange', 'pear']
plt.pie([15, 30, 45, 10], labels=labele, autopct='%1.1f%%', startangle=140)
# here the work of autopct is to show the percentage of each slice in the pie chart
# and the work of startangle is to rotate the pie chart to start at a specific angle
```

```python
labele = ['apple', 'banana', 'orange', 'pear']
sizes = [15, 30, 45,90 ]
plt.pie(sizes, labels=labele, autopct='%1.1f%%', startangle=140)
```

```python
labele = ['Apple', 'Banana', 'Date', 'Grapes']
sizes = [15,30, 45,15]
plt.pie(sizes, labels=labele,autopct='%1.1f%%')
```

```python
from matplotlib import pyplot as plt
players = ['Virat Kohli', 'Rohit Sharma', 'MS Dhoni', 'Sachin Tendulkar']
runs = [5000, 6000, 7000, 8000]
plt.bar(players, runs, color='green')
plt.title('Runs Scored by Players')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```

```python
from matplotlib import pyplot as plt
players = ['Virat Kohli', 'Rohit Sharma', 'MS Dhoni', 'Sachin Tendulkar']
runs = [5000, 6000, 7000, 8000]
plt.barh(players, runs, color='green')
plt.title('Runs Scored by Players')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```
